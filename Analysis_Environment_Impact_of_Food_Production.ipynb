```
import numpy as np
dict_embeddings={}
for product,other_features in zip(df['Food product'],df.iloc[:,1:].to_numpy()):
  pom_list = model_embeddings.wv[product].tolist()
  pom_list.extend(other_features.tolist())
  dict_embeddings[product]=pom_list
```

#<font color='orange'>Clustering</font>

```
from sklearn.preprocessing import StandardScaler
X=StandardScaler().fit_transform(list(dict_embeddings.values()))
```

```
from sklearn.cluster import KMeans

inertia = []
for k in range(1, 15):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(X)
    inertia.append(np.sqrt(kmeans.inertia_))
```

```
plt.plot(range(1, 15), inertia, marker='s')
plt.xlabel('$k$')
plt.ylabel('$J(C_k)$')
```

From the graph we can see that the optimal number of clusters is 3, but from later visualizations we decided that the optimal number of clusters for the dataset is 4.

```
X=[]
y=[]
for embedding in dict_embeddings.values():
  X.append(embedding[:-1])
  y.append([embedding[-1]])

X = pd.DataFrame(X)
y=pd.DataFrame(y)
```

```
from sklearn.model_selection import LeaveOneOut
```

```
LOOCV = LeaveOneOut()

LOOCV.get_n_splits(X)
```

```
for train_index, test_index in LOOCV.split(X):
   print("TRAIN:", train_index, "TEST:", test_index)
   X_train, X_test = X.loc[train_index], X.loc[test_index]
   y_train, y_test = y.loc[train_index], y.loc[test_index]
   print(X_train, X_test, y_train, y_test)
```

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
```

```
from sklearn.metrics import mean_squared_error
```

```
y_true, y_pred = list(), list()

for train_index, test_index in LOOCV.split(X):
    # split the data
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y.loc[train_index], y.loc[test_index]
    # fit the model
    model = LinearRegression()
    model.fit(X_train, y_train)
    # predict y_test
    y_predictions = model.predict(X_test)
    # append to y_true and y_pred
    proba = str(y_test).split()
    #print('Y_true: ', float(proba[1]), ' y_pred: ', round(y_predictions[0], 1))
    y_true.append(float(proba[1]))
    y_pred.append(y_predictions[0])
# evaluate the model
mse = mean_squared_error(y_true, y_pred)
print('MSE %f' % mse)
```

```
from sklearn.model_selection import cross_val_score
from numpy import mean
from numpy import std
from numpy import absolute

cv = LeaveOneOut()
model = LinearRegression()
scores = cross_val_score(model, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
scores = absolute(scores)
print('MAE: %.3f (%.3f)' % (mean(scores), std(scores)))
scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=cv, n_jobs=-1)
scores = absolute(scores)
print('Mean MSE: %.3f (%.3f)' % (scores.mean(), scores.std()) )
```
